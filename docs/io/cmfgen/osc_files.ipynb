# Energy levels and oscillator strengths

```python
from carsus.io.cmfgen import CMFGENEnergyLevelsParser, CMFGENOscillatorStrengthsParser
```

Download and extract CMFGEN Atomic Data from http://kookaburra.phyast.pitt.edu/hillier/web/CMFGEN.htm

```python
si2_energy_levels = CMFGENEnergyLevelsParser('./atomic/SIL/II/16sep15/si2_osc_kurucz')
```

The resulting table is stored as a DataFrame in the *base* attribute. Header information is saved as a dictionary in *meta* attribute.

```python
si2_energy_levels.meta
```

```python
si2_energy_levels.base
```

Same goes for the oscillator strengths table.

```python
si2_osc_strengths = CMFGENOscillatorStrengthsParser('./atomic/SIL/II/16sep15/si2_osc_kurucz')
```

```python
si2_osc_strengths.base
```

You can easily dump the data inside an HDF5 file. For example:

```python
si2_energy_levels.to_hdf()
si2_energy_levels.to_hdf()
```

will result in a `si2_osc_kurucz.h5` file located in the same path and containing both tables and header information.
