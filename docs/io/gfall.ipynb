Import the `GFALLReader` class and load the file.

```python
from carsus.io.kurucz import GFALLReader

gfall_reader = GFALLReader('gfall.dat')
```

```python
gfall_reader.gfall_raw
```

Dump the data with the `to_hdf` method.

```python
gfall_reader.to_hdf('gfall.h5', raw=False)
```
